##  Import libraries 

```python
# Import libraries
import pandas as pd
import numpy as np
import random
```

### 1. Examining and loading data


```python
# Load the data
df1 = pd.read_csv('S3749857_dataset1_solution.csv')
df2 = pd.read_csv('S3749857_dataset2.csv')
# Check the structure of the dataset 2
df2
```

Then, check if there is any missing value in each column of df2.

```python
df2.count()
```

There are missing values in the `Organisation`, `Contract Type` and `Full-Time Equivalent (FTE)` column. We need to solve this problem in the next steps.

### 2. Resolving schema conflicts

#### Conflict 1: missing conflicts in the Id column (e.g. naming conflicts in xx column and xx column)

```python
# Check the columns in df1
print(df1.columns)
# Check the columns in df2
print(df2.columns)
```

There is no such column in df2 which correspond the ID column in df1, so we need to create a new column called ID in df2 and assign 8-digit unique values to the new column. (The values in the new column is also unique in df1)

```python
# Generate unique Id for df2
Id = [] # For storing Id for df2

for var in range(0, df2.shape[0]):
    check = True
    while check:
        ran = random.randint(10000000,99999999)
        # If the randomly produced number is not in the Id column in df1 and df2, pick it as the Id value
        if (ran not in list(df1['Id'])) & (ran not in Id):
            Id.append(ran)
            check = False
            
df2.insert(0, 'Id', Id)
df2
```

#### Conflict 2: naming conflicts in the Title column and Job Title column

```python
# Check the columns in df1
print(df1.columns)
# Check the columns in df2
print(df2.columns)
```

The column `Title` in df1 and `Job Title` is the same thing, but their name is different. We just need to replace the name 'Job Title' in df2 with 'Title'.

```python
# Rename the Job Title column 
df2 = df2.rename(columns = {'Job Title': 'Title'})
df2
```

#### Conflict 3: naming and value conflicts in the Company column and Organisation column

The `Company` column in df1 correspond to the `Organisation` column. We need to rename the `Organisation` column. Besides, there are missing values in the `Organisation` column. We need to replace them with 'non-specified'.

```python
# Rename the Organisation column 
df2 = df2.rename(columns = {'Organisation': 'Company'})
# Replace the missing values with 'non-specified'
df2.loc[df2['Company'].isna(),'Company'] = 'non-specified'
df2
```

#### Conflict 4: naming and value conflicts in the Salary column and Monthly Payment column

The `Salary` column in df1 correspond to the `Monthly Payment` column. We need to rename the `Monthly Payment` column. Moreover, the value is annual salary in the `Salary` column but monthly salary in the `Monthly Payment` column, so we need to multiply the `Monthly Payment` column by 12. Also, we need to convert it with two decimal places.

```python
# Rename the Organisation column 
df2 = df2.rename(columns = {'Monthly Payment': 'Salary'})
# Multiply the Salary column by 12
df2.loc[:,'Salary'] = df2['Salary'] * 12
# Convert the value to the value with two decimal places.
dfCopy = df2.copy() # To avoid SettingWithCopyWarning
dfCopy.loc[:,'Salary'] = df2['Salary'].apply(lambda x: format(x, '.2f'))
df2 = dfCopy
df2
```

#### Conflict 5: naming and value conflicts in the ContractTime column and Contract Type column

```python
# Check the values in the Contract Type column in df2
df2['Contract Type'].value_counts()
```

The `ContractTime` column in df1 correspond to the `Contract Type` column, so we need to rename the `Contract Type` column. Then, in the `Contract Type` column, we also need to change the values 'Permanent' to 'permanent', 'Fixed Term Contract' to 'contract' and missing values to 'non-specified', which will solve the value conflict.

```python
# Rename the Contract Type column 
df2 = df2.rename(columns = {'Contract Type': 'ContractTime'})
# Change the values
df2.loc[df2['ContractTime'] == 'Permanent', 'ContractTime'] = 'permanent'
df2.loc[df2['ContractTime'] == 'Fixed Term Contract', 'ContractTime'] = 'contract'
df2.loc[df2['ContractTime'].isna(), 'ContractTime'] = 'non-specified'
df2['ContractTime'].value_counts()
```

#### Conflict 6: value conflicts in the Category column

```python
# Check the values in the Category column in df2
df2['Category'].value_counts()
```

The values in the `Category` in df2 is similar to that in df1, we need to replace them with the values of the formats in df1 to fix the value conflict.

```python
# Change the values
df2.loc[df2['Category'] == 'Information Technology', 'Category'] = 'IT Jobs'
df2.loc[df2['Category'] == 'Engineering', 'Category'] = 'Engineering Jobs'
df2.loc[df2['Category'] == 'Healthcare and Nursing', 'Category'] = 'Healthcare & Nursing Jobs'
df2.loc[df2['Category'] == 'Finance and Accounting', 'Category'] = 'Accounting & Finance Jobs'
df2.loc[df2['Category'] == 'Sales', 'Category'] = 'Sales Jobs'
df2.loc[df2['Category'] == 'Hospitality and Catering', 'Category'] = 'Hospitality & Catering Jobs'
df2.loc[df2['Category'] == 'Teaching', 'Category'] = 'Teaching Jobs'
df2.loc[df2['Category'] == 'PR, Advertising and Marketing', 'Category'] = 'PR, Advertising & Marketing Jobs'
df2['Category'].value_counts()
```

#### Conflict 7: naming and value conflicts in the ContractTime column and Full-Time Equivalent (FTE) column

```python
# Check the values in the Full-Time Equivalent (FTE) column in df2
df2['Full-Time Equivalent (FTE)'].value_counts()
```

The `ContractType` column in df1 correspond to the `Full-Time Equivalent (FTE)` column, so we need to rename the `Full-Time Equivalent (FTE)` column. Then, in the `Contract Type` column, we also need to change the values 'FULL TIME' to 'full_time’', '0.2 FTE', '0.4 FTE', '0.6 FTE', '0.8 FTE', to 'part_time' and missing values to 'non-specified', which will solve the value conflict.

```python
# Rename the Full-Time Equivalent (FTE) column 
df2 = df2.rename(columns = {'Full-Time Equivalent (FTE)': 'ContractType'})
# Change the values
df2.loc[df2['ContractType'] == 'FULL TIME', 'ContractType'] = 'full_time'
df2.loc[df2['ContractType'] == '0.2 FTE', 'ContractType'] = 'part_time'
df2.loc[df2['ContractType'] == '0.4 FTE', 'ContractType'] = 'part_time'
df2.loc[df2['ContractType'] == '0.6 FTE', 'ContractType'] = 'part_time'
df2.loc[df2['ContractType'] == '0.8 FTE', 'ContractType'] = 'part_time'
df2.loc[df2['ContractType'].isna(), 'ContractType'] = 'non-specified'
df2['ContractType'].value_counts()
```

#### Conflict 8: naming conflicts in the OpenDate column and Opening column, as well as CloseDate column and Closing

The `OpenDate` and `CloseDate` column in df1 correspond to the `Opening` and `Closing` column respectively, so we need to rename the two columns.

```python
# Rename the two columns
df2 = df2.rename(columns = {'Opening': 'OpenDate', 'Closing': 'CloseDate'})
df2
```

#### Conflict 9: missing conflicts in the SourceName column

There is no such column in df2 which correspond the SourceName column in df1, so we need to create a new column called SourceName in df2 and assign 'non-specified' to all values in the new column.

```python
# Create a new column and assign 'non-specified' to all values in the new column
df2.loc[:,'SourceName'] = 'non-specified'
df2
```

### 3. Merging data

The columns and values in df2 are now consistent with those in df1, so we can merge them into one dataframe.

```python
# Merge two data sets
df = pd.concat([df1, df2])
df.index = range(0, df1.shape[0] + df2.shape[0]) # Assign new index to the new data set
df.loc[:,'Salary'] = df['Salary'].astype(float) # Make sure the value in Salary column is in the same format
df
```

### 4. Resolving data conflicts:

#### Conflict 1: complete duplications (e.g., complete duplications)

To check if there is any complete duplication, we should not contain the `Id` and `SourceName` column, because these two columns are the columns we created for merging two dataframes. We need to compare the remaining columns(the original columns in df2) for each row.

```python
# Check if there is any complete duplication in terms of the columns except Id and SourceName
df.duplicated(subset = ['Title','Location','Company','ContractType','ContractTime','Category','Salary','OpenDate','CloseDate']).sum()
```

The result shows that there are 105 duplications in the new dataset. They are apparently redundant, so we need to delete them.

```python
# Delete duplications
df = df[~df.duplicated(subset = ['Title','Location','Company','ContractType','ContractTime','Category','Salary','OpenDate','CloseDate'])]
df
```

#### Finding global key for the data
For the global key, the `Id` columns is the best choice, because the value of Id of each row is unique. We can check if they are unique in case something wrong happening.

```python
# Code to identify global key
df.duplicated(subset = ['Id']).sum()
```

As you can see, each value in the `Id` column is unique. We choose the `Id` column as the global key for the dataset.

### 5. Saving the integrated and reshaped data
The last part of the integration process is to export our output data to csv format, named as:

```python
# Save output data
df.to_csv('S3749857_dataset_integrated.csv', index = False)
```
